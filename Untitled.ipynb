```python
import pandas as pd
import json
from sklearn import preprocessing
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from sklearn.feature_extraction.text import CountVectorizer
```

```python
with open('data_and_analysis/dc_characters_complete.txt') as data_file:    
    data = json.load(data_file)
    
with open('data_and_analysis/marvel_characters_complete.txt') as data_file:    
    data = data + json.load(data_file)
```

```python
powers = {}
origins = {}
names = []

for character in data:
    for power in character['powers']:
        powers[power['id']] = power['name']
    if character['origin']:
        origins[character['origin']['id']] = character['origin']['name']
    names.append(character['name'])    

power_binarizer = preprocessing.LabelBinarizer()
power_binarizer.fit(list(powers.values()))

origin_binarizer = preprocessing.LabelBinarizer()
origin_binarizer.fit(list(origins.values()))


name_vectorizer = CountVectorizer(min_df=.001)
name_vectorizer.fit(names)

#lb.transform(['Magnetism', 'Implants']).sum(axis=0)
```

```python
characters = {}
for character in data:
    if (len(character['powers']) > 0) and character['origin'] and character['gender'] in [1,2]:
        characters[character['name']] = {
            'name': character['name'],
            'gender': character['gender'],
            'powers': power_binarizer.transform([power['name'] for power in character['powers']]).sum(axis=0),
            'origin': origin_binarizer.transform([character['origin']['name']]).reshape(-1),
            'name_vector': name_vectorizer.transform([character['name']]).toarray().reshape(-1)
        }
```
